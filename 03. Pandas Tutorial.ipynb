```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                    [3, 4, np.nan, 1],
                    [np.nan, np.nan, np.nan, 5],
                    [np.nan, 3, np.nan, 4]],
                    columns=list('ABCD'))
```

```python
#Data Frame Created
df
```

```python
#Fill NAs with 0's
df.fillna(0)
```

```python
#Original Dataframe
df
```

```python
#Fill data with specific values
values = {'A': 10, 'B': 11, 'C': 24, 'D': 32}
df.fillna(value=values)
```

```python
#Limit the number of rows to fill the NAs in the dataset
df.fillna(value=values, limit=2)
```

```python
df1 = pd.DataFrame({"name": ['Rahul', 'Akash', 'Vikram'],"age": [22, 27, 32]})
df1
```

```python
df2 = pd.DataFrame({"name": ['Vikram','Rahul', 'Akash'],"education": ["Engineering", "Doctor", "MBA"]})
df2
```

```python
df1.merge(df2,left_on="name",right_on="name",how="left")
```

```python
df3 = pd.DataFrame({"name": ['Vikram','Rahul'],"education": ["Engineering", "Doctor"]})
df3
df1.merge(df3,left_on="name",right_on="name",how="inner")
```

```python
df3 = pd.DataFrame({"name": ['Vikram','Rahul'],"education": ["Engineering", "Doctor"]})
df3
df1.merge(df3,left_on="name",right_on="name",how="left")
```

```python
df3 = pd.DataFrame({"name": ['Vikram','Rahul'],"education": ["Engineering", "Doctor"]})
df3
df1.merge(df3,left_on="name",right_on="name",how="right")
```
